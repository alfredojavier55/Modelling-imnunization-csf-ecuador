```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading CSV
# Observed
Fig24 = pd.read_csv("Dados_2014_betaMJUN.csv", header=None, sep=";")
Fig24.columns = ["Tempo", "Casos"]
# Adjusted
adjusted = pd.read_csv("t.csv")
```

```python
# Plotting and saving the file
plt.figure(figsize=(6,4), dpi=400)
sns.lineplot(x="tempo", y="imode", data=adjusted, color="red", label="Adjusted")
sns.scatterplot(x="Tempo", y="Casos", data=Fig24, color="black", s=15, label="Observed")
plt.xlabel("Time (month)")
plt.ylabel("Number of cases 2014")
plt.legend(title=None)
plt.savefig("beta_gamma14-17.tif", dpi=200, bbox_inches="tight")
plt.show()
```
