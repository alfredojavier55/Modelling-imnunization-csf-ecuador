```python
import pandas as pd
import numpy as np
```

```python
#Lendo o csv, com a primeria coluna como index e sem cabeçalho
Fig24 = pd.read_csv("Dados_2014_betaMJUN.csv", sep=';', index_col=0,header=None)
#Adicionando 'Cases' como nome do cabeçalho, já que só temos uma coluna, com valores de casos
Fig24.columns = ['Cases']
#Mudando o nome do index, que são 12 meses, para 'Months'
Fig24.index.name = 'Months'
```

```python
#Vendo se o df foi importado corretamente
Fig24
```

```python
adjusted = pd.read_csv("t.csv", index_col=0)
adjusted
```

```python
#Instalando e importando as bibliotecas de gráficos
#!pip install seaborn
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
# Alterando o índice de Fig24 para meses
months = ['Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec',
          'Jan', 'Feb', 'Mar', 'Apr', 'May']
Fig24.index = months[:len(Fig24)]

# Criando a figura e os eixos
fig, ax = plt.subplots(figsize=(8, 6))

# Adicionando pontos de Fig24
ax.scatter(Fig24.index, Fig24['Cases'], color='black', label='Observed', s=10)

# Adicionando linha de adjusted
ax.plot(adjusted['tempo'], adjusted['imode'], color='red', label='Adjusted')

# Adicionando rótulos dos eixos
ax.set_xlabel('Time (month)')
ax.set_ylabel('Number of cases 2014')

# Alterando os valores do eixo x, rotacionado para caber
ax.set_xticks(range(len(Fig24)))
ax.set_xticklabels(Fig24.index)

# Adicionando grade, com tipo de linha, transparência e sobreposição
ax.grid(True, linestyle='--', alpha=0.5, zorder=0)

# Adicionando legenda, cujos valores estão especificados no argumento label de cada função de gráfico
ax.legend()

# Adicionado título abaixo do gráfico, fonte 10 a esquerda
fig.suptitle('Fig 24. Adjusted Beta and Gamma', fontsize=10, x=0.05, y=-0.05, ha='left')

# Salvando e visualizando
plt.savefig('Fig24_beta_gamma14-17.png', dpi=300, bbox_inches='tight')
plt.show()
```
